```python
import requests
from bs4 import BeautifulSoup
import time
import random
import pandas as pd
import re
from unidecode import unidecode
import pickle
```

#Collecter des commentaires du produit d’amazone
## Produit << Fire TV Stick Lite avec télécommande vocale Alexa >>

```python
head={"Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
"User-Agent":"Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:88.0) Gecko/20100101 Firefox/88.0"}
```

```python
label=[]
commentaire=[]
for i in range(1,200):
    # Pause avant chaque requete d'un temps arbitraire entre 0.5 et 1.5 s
    time.sleep(random.uniform(0.5,1.5))
    url='https://www.amazon.fr/decouvrez-fire-tv-stick-lite-avec-telecommande-vocale-alexa-lite-sans-boutons-de-controle-de-la-tv-modele-2020/product-reviews/B07ZZVWB4L/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber='+str(i)
    r=requests.get(url,headers=head)
    # Si le retour request est diff de 200 ==> pb
    print(url,r.status_code)
    soup=BeautifulSoup(r.content)

    for elem in soup.findAll('i',attrs={"data-hook":"review-star-rating"}):
        label.append(elem.text.split(',')[0])

    for elem in soup.findAll('span',attrs={"data-hook":"review-body"}):
        commentaire.append(elem.text.strip())

    print("iteration : ",i,"/ longueur liste ",len(label),len(commentaire))
    
    if (len(label),len(commentaire)) != (i*10,i*10):
        m=min(len(label),len(commentaire))
        label=label[:m]
        commentaire=commentaire[:m]
        break
```

```python
data = pd.DataFrame({'Commentaire':commentaire,'Label':label})
```
